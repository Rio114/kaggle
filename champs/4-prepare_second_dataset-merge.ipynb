```python
import pandas as pd
import pickle
```

```python
FOLDER = '../../data_kaggle/champs/'
OUTPUT = FOLDER + 'out/'
```

```python
temp_pickle = '20190721_features_{}.pickle'
temp_pickle_val = '20190721_features_val_{}.pickle'
temp_pickle_sub = '20190721_features_sub_{}.pickle'

second_train = "20190721_second_train.pickle"
second_val = "20190721_second_val.pickle"
second_test = "20190721_second_test.pickle"
```

```python
df_train = pd.read_csv(FOLDER+'train.csv')
df_test = pd.read_csv(FOLDER+'test.csv')
```

```python
df_second_train_list = []
df_second_val_list = []
df_second_test_list = []

for i in range(8):
    df_second_train_list.append(pd.read_pickle(OUTPUT + temp_pickle.format(i)))
    df_second_val_list.append(pd.read_pickle(OUTPUT + temp_pickle_val.format(i)))
    df_second_test_list.append(pd.read_pickle(OUTPUT + temp_pickle_sub.format(i)))
    
df_second_train = pd.concat(df_second_train_list, axis=0)
df_second_val = pd.concat(df_second_val_list, axis=0)
df_second_test = pd.concat(df_second_test_list, axis=0)
```

```python
df_second_train_merge = pd.merge(df_train, df_second_train, on=['id'], how='inner')
df_second_val_merge = pd.merge(df_train, df_second_val, on=['id'], how='inner')
df_second_test_merge = pd.merge(df_test, df_second_test, on=['id'], how='inner')
```

```python
df_second_train_merge.to_pickle(OUTPUT + second_train)
df_second_val_merge.to_pickle(OUTPUT + second_val)
df_second_test_merge.to_pickle(OUTPUT + second_test)
```
